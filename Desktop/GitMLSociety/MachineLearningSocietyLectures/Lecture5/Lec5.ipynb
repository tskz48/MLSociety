```python

import nnfs
import pandas as pd
from nnfs.datasets import spiral_data
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
import torch
import torch.nn as nn # Neural Networks
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import math
import copy

nnfs.init()
X,y=spiral_data(samples=500,classes=2) #number of data points per class is 500. Neural networks are suitable to model complex non-linear patterns in spiral datasets.
#print(X.shape)
print(y)
plt.scatter(X[:,0],X[:,1],c=y,s=6) #c represents colour of the points and s represents the size of the points. c=y means that the points are coloured according to the class (y)
plt.title("Spiral Dataset") 
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
           
plt.show()

scalar=StandardScaler()
X=scalar.fit_transform(X)
X_train,X_test,y_train,y_test= train_test_split(X,y,train_size=0.7,random_state=1)


X_train.shape
y_train=pd.get_dummies(y_train) #one hot encoding of labels (to convert categorical data into numeric values in binary format)
y_test=pd.get_dummies(y_test)
def add_random_features(data,n_features=200):
    random_features=np.random.normal(0,1,size=(data.shape[0],n_features))
    return np.column_stack([data,random_features])
X_train=add_random_features(X_train)
X_test=add_random_features(X_test)
print(X_train.shape) #now the size of the feature space increased to 202. We added 200 features

#Question 1
inputs=tf.keras.Input(shape=(202,)) #202 input features
hidden=tf.keras.layers.Dense(512,activation='relu')(inputs) #first hidden layer with 512 neurons, and we apply relu activation on the weighted sum: sum(weights*inputs) + bias.
hidden=tf.keras.layers.Dense(256,activation='relu')(hidden) #second hidden layer with 256 neurons. We apply relu activation on the weighted sum.
outputs=tf.keras.layers.Dense(2,activation='softmax')(hidden) #output layer (2 neurons, one per class) and a softmax activation.

tf_model=tf.keras.Model(inputs,outputs)

tf_model.compile(
   optimizer='adam',#adaptive learning rate optimisation
   loss = 'binary_crossentropy', #we used binary_crossentropy loss because y_train is one-hot encoded.
   metrics=['accuracy'])

history = tf_model.fit(
    X_train,
    y_train,
    validation_split=0.12, #12% of the dataset is used for validation and the other 88% is used for training.
    batch_size=32, #processes 32 samples at a time before updating the weights.
    epochs=500  #the complete training dataset is passed through the machine learning training algorithm 500 times.
 )

tf_rmse=np.sqrt(tf_model.evaluate(X_test,y_test))
print(tf_rmse)

class spiralclassifier(nn.Module): #pytorch implementation
    def __init__(self):
        super(spiralclassifier, self).__init__()
        self.layer1 = nn.Linear(202, 512) 
        self.layer2 = nn.Linear(512, 256)
        self.output = nn.Linear(256, 2)
        
    def forward(self, x):
        x = F.relu(self.layer1(x)) #relu activation function, defined by relu(x)=max(0,x), so converts all negative values to 0. Relu is good to prevent the vanishing gradient problem, since the derivative is 1 for positive inputs.
        x = F.relu(self.layer2(x))
        x = F.softmax(self.output(x)) #converts to probabilities ranging from 0 to 1. The class with the highest probability is the predicted label.
        
        return x


t_X_train = torch.tensor(X_train).type(torch.float32) #converts X_train to a pytorch tensor, which is of floating point 32-bit format.
t_y_train = torch.tensor(np.array(y_train)).type(torch.float32) #we had to convert y_train from a pandas dataframe to a numpy array, to make it compatible with torch.tensor(). X_train was already a numpy array because we applied scalar.fit_transform prior.

t_X_test = torch.tensor(X_test).type(torch.float32)  #converts X_test to a pytorch tensor
t_y_test = torch.tensor(np.array(y_test)).type(torch.float32)
model = spiralclassifier()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01) # Use the Adam Optimizer
criterion = nn.BCELoss() #Binary cross entropy loss criterion.

# train_dataset = TensorDataset(t_X_train, t_y_train)
# train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True) #shuffle = True randomises the order of data, which ensures the model does not memorise.

# #we incorporate mini-batch processing to increase computational efficiency than processing each sample individually
# # for epoch in range(1, 501):  # Include the last epoch
# #     total_loss = 0  # Reset loss for the epoch
    
# #     for batch_X, batch_y in train_loader:
# #         optimizer.zero_grad()
# #         output = model(batch_X)
# #         loss = criterion(output, batch_y)
# #         loss.backward() #backpropogation.
# #         optimizer.step() #updates weights

# #         total_loss += loss.item()
    
# #     print(f"Epoch {epoch}, Loss: {total_loss/len(train_loader)}")  




for x, target in zip(t_X_train, t_y_train):
    optimizer.zero_grad()#resets the gradient of all model parameters to zero, so gradients from the previous iteration do not accumulate in the next iteration, which leads to less accurate parameter updates and hence training.
    output = model(x) #forward pass
    loss = criterion(output, target)
    loss.backward() #backpropagation, computes the gradient of the loss with respect to the model parameters, for each layer.
    optimizer.step()#updates model parameters

total_loss = 0

# Testing Model with validation data

for x, target in zip(t_X_test, t_y_test):
    output = model(x)
    loss = criterion(output, target)
    total_loss += loss
    
avg_loss = total_loss / len(t_X_test) #average binary cross entropy loss for the model

torch_rmse = torch.sqrt(torch.tensor(avg_loss)).detach().numpy() #detach ensures no gradient information is stored and tracked for the tensor anymore, which can saves memory and computational resources. Also, when computing metrics such as RMSE, gradients need not be computed.

print("TensorFlow RMSE:", tf_rmse)
print("PyTorch RMSE:", torch_rmse)



# test_dataset = TensorDataset(t_X_test, t_y_test)
# test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)  

# total_loss = 0

# # Set model to evaluation mode (disables dropout, batch norm updates)
# model.eval()

# with torch.no_grad():  # Disable gradient tracking for efficiency
#     for batch_X, batch_y in test_loader:
#         output = model(batch_X)  # Forward pass
#         loss = criterion(output, batch_y)  # Compute loss
#         total_loss += loss.item() * batch_X.size(0)  # Sum loss over batch

# # Compute average loss
# avg_loss = total_loss / len(t_X_test)
# torch_rmse = torch.sqrt(torch.tensor(avg_loss)).numpy()  # Convert to NumPy for readability

# print("PyTorch RMSE:", torch_rmse)





# #for each epoch, we compute accuracy, loss, validation accuracy and validation loss.
```

```python
#Question2

# Dataset with two inputs and one target output

import numpy as np

# Assuming spiral_data() returns X and y
X, y = spiral_data(samples=500, classes=2)

# Select two specific samples (rows)
sample1 = X[0,:]  # First sample
sample2 = X[1,:]  # Second sample
target1 = y[0]
target2=y[1]
# Stack them together into a NumPy array
samples_array = np.vstack([sample1, sample2])
target_array=np.array([target1, target2])
print(target_array)
# X = np.array([[1, 2], [2, 3], [3, 4], [4, 5]])  # Inputs (2 features per example)
# y = np.array([3, 5, 7, 9])  # Targets (1 output per example)

# Initial weights and bias

w_init = np.random.randn(samples_array.shape[1]) * 0.01 
#w_init = np.zeros(samples_array.shape[1])  # Initialize weights (shape matches number of features)
b_init = 0.0                   # Initialize bias

# ReLU activation function and its derivative
def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return (x > 0).astype(float) # (x>0) True if x>0, otherwise False. We convert to float, so True is given a value 1.0 and False is given a value 0.0.



class Layer_Dense: #represented fully connected layer of the neural network where each neurone is connected to every neuron in the previous layer. The weighted sum of inputs is computed and the bias is added to the sum

    # Layer initialization
    def __init__(self, n_inputs, n_neurons):
        # Initialize weights and biases
        self.weights = 0.01 * np.random.randn(n_inputs, n_neurons)
        self.biases = np.zeros((1, n_neurons))

    # Forward pass
    def forward(self, inputs):
        # Calculate output values from inputs, weights and biases
        self.output = np.dot(inputs, self.weights) + self.biases


class Activation_ReLU: 

    # Forward pass
    def forward(self, inputs):
        # Calculate output values from inputs
        self.output = np.maximum(0, inputs)

class Activation_Softmax: #helps to predict class probabilities

    # Forward pass
    def forward(self, inputs):

        # Get unnormalized probabilities
        exp_values = np.exp(inputs - np.max(inputs, axis=1,
                                            keepdims=True))
        # Normalize them for each sample
        probabilities = exp_values / np.sum(exp_values, axis=1,
                                            keepdims=True)

        self.output = probabilities

class Loss:

    # Calculates the data and regularization losses
    # given model output and ground truth values
    def calculate(self, output, y):

        # Calculate sample losses
        sample_losses = self.forward(output, y)
        

        # Calculate mean loss
        data_loss = np.mean(sample_losses)

        # Return loss
        return data_loss

class Loss_CategoricalCrossentropy(Loss): #we use CategoricalCrossEntropy as the loss function for multi-class classification, where there are more than two classes. We have three different classe in our case.
                                          #the Loss_CategoricalCrossentropy class inherits from the Loss class. In particular, the method 'calculate' is inherited by categorical_cross_entropy class. This categorical_cross_entropy is an extension of the Loss class, as it adds further functionality with its forward method.
  
    # Forward pass
    def forward(self, y_pred, y_true):

        # Number of samples in a batch
        samples = len(y_pred)

        # Clip data to prevent division by 0
        # Clip both sides to not drag mean towards any value
        y_pred_clipped = np.clip(y_pred, 1e-7, 1 - 1e-7)


        # Probabilities for target values -
        # only if categorical labels
        if len(y_true.shape) == 1:
            correct_confidences = y_pred_clipped[
                range(samples),
                y_true
            ]

        # Mask values - only for one-hot encoded labels
        elif len(y_true.shape) == 2:
            correct_confidences = np.sum(
                y_pred_clipped*y_true,
                axis=1
            )

        # Losses
        negative_log_likelihoods = -np.log(correct_confidences)
        return negative_log_likelihoods


dense1 = Layer_Dense(2, 3)

# Create ReLU activation (to be used with Dense layer):
activation1 = Activation_ReLU()

# Create second Dense layer with 3 input features (as we take output
# of previous layer here) and 3 output values
dense2 = Layer_Dense(3, 3)

# Create Softmax activation (to be used with Dense layer):
activation2 = Activation_Softmax()

# Create loss function
loss_function = Loss_CategoricalCrossentropy()

# Perform a forward pass of our training data through this layer
dense1.forward(X)

# Perform a forward pass through activation function
# it takes the output of first dense layer here
activation1.forward(dense1.output)


# Perform a forward pass through second Dense layer
# it takes outputs of activation function of first layer as inputs
dense2.forward(activation1.output)

# Perform a forward pass through activation function
# it takes the output of second dense layer here
activation2.forward(dense2.output)

# Let's see output of the first few samples:
print(activation2.output[:10])

# Perform a forward pass through loss function
# it takes the output of second dense layer here and returns loss
loss = loss_function.calculate(activation2.output, y)

# Print loss value


# Cost function (Mean Squared Error after ReLU)
def compute_cost(X, y, w, b):
    m = X.shape[0]
    z = np.dot(X, w) + b  # Linear part
    predictions = relu(z)  # Apply ReLU
    cost = (1 / (2 * m)) * np.sum((predictions - y) ** 2)
    return cost

# Gradient function
def compute_gradient(X, y, w, b):
    m, n = X.shape
    dj_dw = np.zeros((n,))
    dj_db = 0.0

    for i in range(m):
        z = np.dot(X[i], w) + b  # Linear output
        a = relu(z)  # Apply ReLU
        err = a - y[i]  # Error

        # Backpropagation through ReLU
        relu_grad = relu_derivative(z)  # Derivative of ReLU
        err *= relu_grad  # Chain rule adjustment for ReLU

        for j in range(n):
            dj_dw[j] += err * X[i, j]
        dj_db += err

    dj_dw/= m
    dj_db/= m
    return dj_db, dj_dw

# Gradient Descent function
def gradient_descent(X, y, w_in, b_in, cost_function, gradient_function, alpha, num_iters):
    J_history = []
    w = copy.deepcopy(w_in)#changes made to the deepcopied object do not affect the original object.
    b = b_in

    for i in range(num_iters):
        # Compute gradients
        dj_db, dj_dw = compute_gradient(X, y, w, b)
        
        # Update parameters
        w = w - alpha * dj_dw
        b = b - alpha * dj_db
        
        # Save cost
        if i < 100000:
            J_history.append(cost_function(X, y, w, b))
        
        # Print cost at intervals
        if i % math.ceil(num_iters / 10) == 0:
            print(f"Iteration {i:4d}: Cost {J_history[-1]:8.4f}")
    
    return w, b, J_history


def zscore_normalize_features(X):
    mu = np.mean(X, axis=0)                
    sigma = np.std(X, axis=0)                  
    X_norm = (X - mu) / sigma      

    return (X_norm, mu, sigma)
# normalize the original features
X_norm, X_mu, X_sigma = zscore_normalize_features(samples_array)


# Gradient Descent Parameters
alpha = 0.01       # Learning rate
num_iters = 1000   # Number of iterations

# Train the neural network using gradient descent with ReLU
w, b, J_history = gradient_descent(X_norm, target_array, w_init, b_init, compute_cost, compute_gradient, alpha, num_iters)

# Final Parameters
print("\nTrained Parameters:")
print(f"Weights: {w}")
print(f"Bias: {b}")

# Test the trained model
X_test = np.array([[5, 6], [6, 7]])  # New input examples
z_test = np.dot(X_test, w) + b  # Linear output
predictions = relu(z_test)  # Apply ReLU
print("\nPredictions for test inputs:")
print(predictions)
```
